```
import pandas as pd
df = pd.read_csv('apple_products.csv')
df.head()
```

```
# Checking the shape of the DataFrame
print('Shape of the DataFrame:', df.shape)

# Checking the data types of the columns
print('\nData types of the columns:')
print(df.dtypes)

# Checking for missing values
print('\nMissing values in each column:')
print(df.isnull().sum())
```

```
df.describe(include='all')
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style of the plots
sns.set()

# Creating a histogram for the 'Sale Price' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Sale Price'], bins=30, kde=True)
plt.title('Distribution of Sale Price')
plt.show()

# Creating a boxplot for the 'Sale Price' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Sale Price'])
plt.title('Boxplot of Sale Price')
plt.show()

# Creating a bar plot for the 'Discount Percentage' column
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Discount Percentage'])
plt.title('Count of each Discount Percentage')
plt.xticks(rotation=90)
plt.show()
```

```
# Creating a new column 'Popularity' as the sum of 'Number Of Ratings' and 'Number Of Reviews'
df['Popularity'] = df['Number Of Ratings'] + df['Number Of Reviews']

# Sorting the DataFrame based on the 'Popularity' column
popular_products = df.sort_values('Popularity', ascending=False)

# Displaying the top 10 most popular products
popular_products[['Product Name', 'Popularity']].head(10)
```

```
# Creating a bar chart for the top 10 most popular products
top_10_popular = popular_products[['Product Name', 'Popularity']].head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Popularity', y='Product Name', data=top_10_popular, palette='viridis')
plt.title('Top 10 Most Popular Apple Products')
plt.xlabel('Popularity')
plt.ylabel('Product Name')
plt.show()
```

```
# Creating a histogram for the 'Discount Percentage' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Discount Percentage'], bins=30, kde=True)
plt.title('Distribution of Discount Percentage')
plt.show()

# Creating a scatter plot to see if there's a correlation between 'Discount Percentage' and 'Popularity'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount Percentage', y='Popularity', data=df)
plt.title('Discount Percentage vs Popularity')
plt.show()
```

```
# Creating a histogram for the 'Number Of Ratings' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Number Of Ratings'], bins=30, kde=True)
plt.title('Distribution of Number Of Ratings')
plt.show()

# Creating a histogram for the 'Number Of Reviews' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Number Of Reviews'], bins=30, kde=True)
plt.title('Distribution of Number Of Reviews')
plt.show()

# Creating a scatter plot to see if there's a correlation between 'Number Of Ratings' and 'Number Of Reviews'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number Of Ratings', y='Number Of Reviews', data=df)
plt.title('Number Of Ratings vs Number Of Reviews')
plt.show()
```

After performing the Exploratory Data Analysis (EDA) on the Apple products dataset, I have gained several insights:

1.Data Overview: The dataset contains 62 rows and 11 columns. There are no missing values in the dataset. The data types of the columns are appropriate for the data they represent.

2.Descriptive Statistics: The average sale price of the Apple products is approximately 80,073.9. The average discount percentage is approximately 9.95%. The average number of ratings is approximately 22,420.4. The average number of reviews is approximately 1861.68.

3.Product Popularity: The iPhone SE models and the iPhone XR models are the most popular based on the number of ratings and reviews.

4.Price Distribution: The majority of the products are priced between 40,000 and 80,000, with a few products priced higher.

5.Discount Analysis: A large number of products have a discount percentage of around 0 to 10%. There doesn't seem to be a clear correlation between the discount percentage and the popularity of the products.

6.Ratings and Reviews Analysis: A large number of products have a small number of ratings and reviews, while a few products have a very large number of ratings and reviews. There is a positive correlation between the number of ratings and the number of reviews.
